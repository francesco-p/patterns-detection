# Dataset visualization Notebook

```python
import matplotlib.pyplot as plt 
import numpy as np
import cv2
```

## Class ground truth 

- **semantic_lvl1**: Every products made by the same manufacturer share same class.
- **semantic_lvl2**: Equal products share the same label.

```python
fig, axs = plt.subplots(nrows=1, ncols=3, figsize=(15,5))
img = cv2.cvtColor(cv2.imread("../datasets/semantic_lvl1/0118_01.png"), cv2.COLOR_RGB2BGR)
lvl1 = np.load("../datasets/semantic_lvl1/annotation/SegmentationClass/0118_01.npy")
lvl2 = np.load("../datasets/semantic_lvl2/annotation/SegmentationClass/0118_01.npy")

for i, (ax, matrix, t) in enumerate(zip(axs, [img, lvl1, lvl2], ["Original Img", "semantic_lvl1", "semantic_lvl2"])):
    axs[i].imshow(matrix)
    axs[i].set_title(t)
    axs[i].axis('off')
plt.show()
```

## Object ground truth 

Every object has its own label, useful to distinguish different objects and to compute metrics, we report only semantic_lvl1).

```python
gt_obj = np.load("../datasets/semantic_lvl1/annotation/SegmentationObject/0118_01.npy")
plt.figure(figsize=(5,5))
plt.imshow(gt_obj)
plt.axis('off')
plt.title("Obj ids for lvl1")
plt.show()
```
